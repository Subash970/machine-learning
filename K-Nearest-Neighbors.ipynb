```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
columns = ['fLength' , 'fWidth' , 'fSize' , 'fConc' , 'fConc1' , 'fAsym' , 'fM3Long' , 'Trans' , 'fAlpha' , 'fDist' , 'class']
df = pd.read_csv("csv-files/magic04.data" , names = columns)
df.head()
```

```python
#convert the class into numerics

df["class"] = (df['class'] == "g").astype(int)
```

```python
df.head()
```

```python
#split the data into X and y for training and testing

X = df.drop('class' , axis = 1)
y = df['class']

from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2)

X_train.shape , y_train.shape , X_test.shape , y_test.shape
```

```python
#knn Algorithm

from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors = 5)

knn_model.fit(X_train , y_train)
```

```python
#predict the values

y_pred = knn_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test , y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
```

```python
lr_model.fit(X_train , y_train)
```

```python
lr_y_pred = lr_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy Score of Logistic Regresison Algorithm :" , accuracy_score(y_test , lr_y_pred))
```
