```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
raw_df = pd.read_csv("csv-files/Social_Network_Ads.csv")
raw_df.head()
```

```python
raw_df.drop("User ID" , axis = 1 , inplace = True)
```

```python
raw_df["Gender"] = (raw_df["Gender"] == "Male").astype(int)
```

```python
df = raw_df
```

```python
#split the  data into X and y

X = df.drop("Purchased", axis =1)
y = df["Purchased"]

print("Shape of X :", X.shape)
print("Shape of y :", y.shape)
```

```python
#split the dataset into train and test

from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X ,y , train_size = 0.8)

print("Shape of Training Dataset :",X_train.shape , y_train.shape)
print("Shape of Testing  Dataset :",X_test.shape , y_test.shape)
```

```python
#Linear Regression Algorithm

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
#train the model using training Dataset

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

model.fit(scaler.fit_transform(X_train) , y_train)
```

```python
#predict the X_test values

y_pred = model.predict(scaler.fit_transform(X_test))
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy Score :" , accuracy_score(y_test , y_pred))
```

```python
x = np.arange(len(y_test))

plt.figure(figsize=(10, 6))
plt.bar(x, y_test, width=0.4, label='Actual', align='center', alpha=0.4)
plt.bar(x, y_pred, width=0.4, label='Predicted', align='edge', alpha=0.7)
plt.xlabel('Samples')
plt.ylabel('Class')
plt.title('Comparison of Actual vs Predicted Values')
plt.legend()
plt.show()
```
