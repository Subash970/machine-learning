```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
columns_title = ["initial payment" , "last payment" , "credit score" , "house number" , "results" ]
df = pd.read_csv("csv-files/Decision_Tree_ Dataset.csv" , names=columns_title)
df.drop(index = 1 , inplace=True)
```

```python
df.drop("house number" , axis = 1 , inplace=True)
```

```python
df.head()
```

```python
#split the Dataset into X and y
X = df.drop("results" , axis = 1)
y = df["results"]

X.shape,y.shape
```

```python
#split the  X and y dataset into train and split

from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2)

X_train.shape , y_train.shape , X_test.shape , y_test.shape
```

```python
#Decision Tree Algorithm

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
```

```python
#train the model using training Set

model.fit(X_train , y_train)
```

```python
#predict the values of X_test set

y_pred = model.predict(X_test)
```

```python
#accuracy score

from sklearn.metrics import accuracy_score
print("Accuracy Score :" , accuracy_score(y_test , y_pred)*100)
```
